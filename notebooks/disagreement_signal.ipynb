# Disagreement as a signal — the MFD read

The PM already computes `panel_disagreement` per meeting (`pm/disagreement.py`) and
uses it only to size down. Bali-Kelly-Mörke-Rahman (*Machine Forecast Disagreement*)
argue the **dispersion across heterogeneous forecasters is itself informative**. This
notebook tests that offline over a saved PM run — no model calls — via
`evaluation.disagreement_signal`.

Four questions:
1. **Uncertainty read** — does disagreement predict the *size* of the next move?
2. **Accuracy conditioning** — is the PM less accurate when the panel is split?
3. **Trade the dispersion** — the signed IC of disagreement against the move.
4. **Graph → disagreement** — does the analyst `missing_inputs` dependency-graph
   density predict a more split panel?

> Read the **t-statistic**, not the third decimal: at ~12 meetings a year a lone
> `|t| < 2` is consistent with noise.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
from src.layered.evaluation.disagreement_signal import evaluate_run, summarize

# import-by-path, not from the package __init__: this evaluator reuses pm_bench's
# clock rebuild and so pulls in the PM layer (see the module docstring).
PM_RUN = '../reports/pm/duration_on.jsonl'
BOARD  = '../reports/ab'
result = evaluate_run(PM_RUN, board_dir=BOARD, board_suffix='_on')
print(summarize(result))
```

## The accuracy-conditioning split (Q2)

Pooled PM calls, median-split on the meeting's disagreement. If the PM is materially
worse on the high-disagreement half, disagreement is a usable trust-discount signal —
the calibrated use the pod mandate already gestures at.

## Reading it

- **Q1/Q3 (magnitude):** `disagreement → |move|` — a positive IC means a split panel
  precedes a bigger move (an uncertainty read), whatever its direction.
- **Q2 (conditioning):** compare `hit_rate` and `ic` across the two halves. A large gap
  is the finding — trust the PM less when the room is split.
- **Q4 (graph):** `graph_density → disagreement`. A positive IC is MFD's structural
  predictor: the more evidence the panel says it lacks, the more it disagrees.

None of this feeds the prompt — it is a diagnostic, held to the honesty rule
(`docs/analyst-layer.md` §6). Whether disagreement should *drive the trade* is the
decision this evidence informs, not one the PM makes for itself.
